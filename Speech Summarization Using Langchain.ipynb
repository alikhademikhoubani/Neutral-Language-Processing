# StuffDocumentChain Text Summarization

```python
loader = Docx2txtLoader("/kaggle/input/speechpersian/speech.docx")
data = loader.load()
data
```

```python
print(data[0].page_content)
```

```python
def text_cleaning(text):
    text = re.sub("\u200c", " ", text)
    text = re.sub("\xa0", " ", text)
    return text.strip()
```

```python
doc = text_cleaning(data[0].page_content)
print(doc)
```

```python
doc
```

```python
docs = [Document(page_content = doc)]
docs
```

```python
template = """
متن زیر را خلاصه کن:
متن: '{text}'
"""

prompt = PromptTemplate(
    template = template,
    input_variables = ['text']
)

chain = load_summarize_chain(
    llm = llm,
    chain_type = 'stuff',
    prompt = prompt,
    verbose = False
)

response = chain.run(docs)
print(response)
```

# Summarizing Large Documents Using Map Reduce

```python
loader = Docx2txtLoader("/kaggle/input/speechpersian2/speech2.docx")
data = loader.load()
data
```

```python
doc = text_cleaning(data[0].page_content)
doc
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    separators = ["\n\n", "\n", " "],
    chunk_size = 2000,
    chunk_overlap = 200
)

chunks = text_splitter.split_text(doc)
print(len(chunks))
```

```python
for chunk in chunks:
    print(len(chunk))
```

```python
documents = [Document(page_content = chunk) for chunk in chunks]
documents
```

```python
chain = load_summarize_chain(
    llm,
    chain_type = 'map_reduce',
    verbose = True
)

response = chain.run(documents)
print(response)
```

# Map Reduce with Custom Prompts

```python
chunk_prompt = """
متن زیر را خلاصه کن:
متن: '{text}'
خلاصه:
"""

map_prompt_template = PromptTemplate(
    template = chunk_prompt,
    input_variables = ['text']
)
```

```python
combine_prompt = """
یک خلاصه نهایی از کل متن با توجه به نکات زیر تولید کن:
یک تیتر به متن اضافه کن،
متن خلاصه شده را با یک مقدمه کوتاه شروع کن.
متن: '{text}'
"""

combine_prompt_template = PromptTemplate(
    template = combine_prompt,
    input_variables = ['text']
)
```

```python
chain = load_summarize_chain(
    llm = llm,
    chain_type = 'map_reduce',
    map_prompt = map_prompt_template,
    combine_prompt = combine_prompt_template,
    verbose = True
)

response = chain.run(documents)
print(response)
```

# RefineChain for Summarization

```python
chain = load_summarize_chain(
    llm = llm,
    chain_type = 'refine',
    verbose = True
)

response = chain.run(documents)
print(response)
```

```python
initial_prompt_template = PromptTemplate(
    input_variables = ['text'],
    template = """
    متن زیر را به طور خلاصه بنویس به طوری که نکات کلیدی آن حفظ شود:
    {text}
    خلاصه:
    """
)
```

```python
refine_prompt_template = PromptTemplate(
    input_variables = ['existing_answer', 'text'],
    template = """
    خلاصه فعلی: 
    {existing_answer}

    متن جدید:
    {text}

    بر اساس متن جدید،، خلاصه را اصلاح و بهبود بده به طوری که نکات کلیدی متن جدید حفظ شوند.
    خلاصه بهبود یافته:
    """
)
```

```python
chain = load_summarize_chain(
    llm = llm,
    chain_type = 'refine',
    question_prompt = initial_prompt_template,
    refine_prompt = refine_prompt_template,
    verbose = True
)

response = chain.run(documents)
print(response)
```
