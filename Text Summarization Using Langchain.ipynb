# StuffDocumentChain Text Summarization

```python
loader = Docx2txtLoader("/kaggle/input/farsimatn/matnfarsi.docx")
data = loader.load()
data
```

```python
data[0].page_content
```

```python
def text_cleaning(text):
    text = re.sub("\u200c", " ", text)
    text = re.sub("\xa0", " ", text)
    return text.strip()
```

```python
doc = text_cleaning(data[0].page_content)
```

```python
doc
```

```python
docs = [Document(page_content = doc)]
docs
```

```python
template = '''یک خلاصه کوتاه از متن زیر تولید کن.
متن: '{text}'
'''
prompt = PromptTemplate(
    input_variables = ['text'],
    template = template
)
```

```python
chain = load_summarize_chain(
    llm,
    chain_type = 'stuff',
    prompt = prompt,
    verbose = False
)
output_summary = chain.run(docs)
print(output_summary)
```

# Summarizing Large Documents Using Map Reduce

```python
loader = Docx2txtLoader("/kaggle/input/article2/art2.docx")
docs = loader.load()
```

```python
docs
```

```python
data = docs[0].page_content
```

```python
data = text_cleaning(data)
```

```python
print(data)
```

```python
llm.get_num_tokens(data)
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    separators = ["\n\n", "\n", " "],
    chunk_size = 1000,
    chunk_overlap = 100
)
chunks = text_splitter.split_text(data)
print(len(chunks))
```

```python
for chunk in chunks:
    print(len(chunk))
```

```python
documents = [Document(page_content = chunk) for chunk in chunks]
documents
```

```python
chain = load_summarize_chain(
    llm,
    chain_type = 'map_reduce',
    verbose = True
)

summary = chain.run(documents)
print(summary)
```

# Map Reduce with Custom Prompts

```python
chunk_prompt = """
لطفا متن زیر را خلاصه کن:
متن: '{text}'
خلاصه:
"""

map_prompt_template = PromptTemplate(
    input_variables = ['text'],
    template = chunk_prompt
)
```

```python
final_combine_prompt = '''
یک خلاصه نهایی از کل متن با توجه به نکات مهم زیر تولید کن.
یک تیتر اضافه کن،
متن خلاصه را با یک مقدمه آغاز کن.
متن: '{text}'
'''

final_combine_prompt_template = PromptTemplate(
    input_variables = ['text'],
    template = final_combine_prompt
)
```

```python
summary_chain = load_summarize_chain(
    llm = llm,
    chain_type = 'map_reduce',
    map_prompt = map_prompt_template,
    combine_prompt = final_combine_prompt_template,
    verbose = True
)

output = summary_chain.run(documents)
print(output)
```

# RefineChain for Summarization

```python
chain = load_summarize_chain(
    llm = llm,
    chain_type = 'refine',
    verbose = True
)

output_summary = chain.run(documents)
print(output_summary)
```
