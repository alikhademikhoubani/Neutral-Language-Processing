```python
from langchain_groq import ChatGroq
from langchain.chains import ConversationChain
from langchain.memory import ConversationBufferWindowMemory
from langchain.agents import AgentType, load_tools, initialize_agent
```

```python
llm = ChatGroq(
    model_name = "deepseek-r1-distill-llama-70b",
    groq_api_key = "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",
    temperature = 0
)
```

```python
memory = ConversationBufferWindowMemory(k = 4)
convo = ConversationChain(
    llm = llm, 
    memory = memory
)
```

```python
response = convo.run("Hi, my name is ali. how are you?")
print(response)
```

```python
response2 = convo.run("What is my name?")
print(response2)
```

```python
response3 = convo.run("What is machine learning?")
print(response3)
```

```python
response4 = convo.run("Tell me more")
print(response4)
```

```python
tools = load_tools(["wikipedia"], llm = llm)

agent = initialize_agent(
    tools,
    llm,
    agent = AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    handle_sparsing_errors = True,
    verbose = True,
    memory = memory
)
```

```python
response = agent.run("متن کوتاهی درباره علی دایی برای من بنویس")
print(response)
```
