```python
from langchain_groq import ChatGroq
from langchain.vectorstores.cassandra import Cassandra
from langchain.indexes.vectorstore import VectorStoreIndexWrapper
from langchain.embeddings import HuggingFaceEmbeddings
import cassio
from langchain_community.document_loaders.word_document import Docx2txtLoader
import re
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
def cleaning_text(text):
    text = re.sub("\u200c", " ", text)
    text = re.sub("\xa0", " ", text)
    return text.strip()
```

```python
def load_doc(file_path):
    loader = Docx2txtLoader(file_path)
    document = loader.load()
    doc = cleaning_text(document[0].page_content)
    return doc
```

```python
llm = ChatGroq(
    model_name = "llama-3.3-70b-versatile",
    groq_api_key = "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",
    temperature = 0
)
```

```python
embeddings = HuggingFaceEmbeddings(model_name = "sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2")
```

```python
astra_vector_store = Cassandra(
    embedding = embeddings,
    table_name = "qa_mini_demo",
    session = None,
    keyspace = None
)
```

```python
def split_texts(document, chunk_size = 200, chunk_overlap = 10):
    text_splitter = RecursiveCharacterTextSplitter(
        separators = ["\n\n", "\n", " "],
        chunk_size = chunk_size,
        chunk_overlap = chunk_overlap
    )
    chunks = text_splitter.split_text(document)
    return chunks
```

```python
texts = split_texts(doc)
print(len(texts))
```

```python
astra_vector_store.add_texts(texts)
astra_vector_index = VectorStoreIndexWrapper(vectorstore = astra_vector_store)
```

```python
first_question = True
while True:
    if first_question:
        query_text = input("\nEnter your question (or type 'quit' to exit): ").strip()
    else:
        query_text = input("\nWhat's your next question (or type 'quit' to exit): ").strip()
    if query_text.lower() == "quit":
        break
    if query_text == "":
        continue
    first_question = False
    print("\nQUESTION: \" %s\"" % query_text)
    answer = astra_vector_index.query(query_text, llm = llm).strip()
    print("ANSWER: \"%s\"\n" % answer)
    print("FIRST DOCUMENT BY RELEVANCE:")
    for doc, score in astra_vector_store.similarity_search_with_score(query_text, k = 4):
        print("        [%0.4f] \"%s ...\"" % (score, doc.page_content[:84]))
```
