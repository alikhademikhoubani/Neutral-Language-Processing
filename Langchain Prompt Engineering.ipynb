```python
demo_template = """I want you to act as a acting financial advisor for people.
In an easy way, explain the basics of {financial_concept}."""
```

```python
prompt_template = PromptTemplate(
    input_variables = ['financial_concept'],
    template = demo_template
)

prompt_template.format(financial_concept = "income tax")
```

```python
first_chain = LLMChain(llm = llm, prompt = prompt_template)
response = first_chain.run({"financial_concept": "income tax"})
print(response)
```

```python
prompt_template = PromptTemplate(
    input_variables = ["sentence", "target_language"],
    template = "In an easy way translate the following sentence '{sentence}' into {target_language}."
)
prompt_template.format(sentence = "How are you?", target_language = "Persian")
```

```python
second_chain = LLMChain(llm = llm, prompt = prompt_template)
response = second_chain({"sentence": "Hello how are you", "target_language": "persian"})
print(response)
```

```python
examples = [
    {"word": "happy", "antonym": "sad"},
    {"word": "tall", "antonym": "short"}
]

example_formatter_template = """Word: {word}
Antonym: {antonym}
"""

example_prompt = PromptTemplate(
    input_variables = ["word", "antonym"],
    template = example_formatter_template
)
```

```python
few_shot_prompt = FewShotPromptTemplate(
    examples = examples,
    example_prompt = example_prompt,
    prefix = "Give the antonym of every input\n",
    suffix = "Word: {input}\nAntonym: ",
    input_variables = ["input"],
    example_separator = "\n"
)
```

```python
print(few_shot_prompt.format(input = "big"))
```

```python
chain = LLMChain(llm = llm, prompt = few_shot_prompt)
response = chain({"input": "big"})
print(response)
```
