```python
from agno.agent import Agent
from agno.tools import tool
from scrapling import Fetcher
from agno.tools.telegram import TelegramTools
```

```python
from agno.models.ollama import Ollama
from agno.models.groq import Groq
import os 
from dotenv import load_dotenv
```

```python
load_dotenv()
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
```

```python
llm = Groq(id = "qwen/qwen3-32b")
```

```python
@tool(name = 'divar', description = 'retrive car prices')
def divar():
    fetcher = Fetcher()
    x = fetcher.get('https://divar.ir/s/tehran/auto', stealthy_headers = True)
    names = x.css('div.kt-post-card__info > h2::text')
    prices= x.css('div.kt-post-card__info > div:nth-child(3)::text')
    km = x.css('div > div.kt-post-card__info > div:nth-child(2)::text')
    cars = list(zip(names, prices, km))
    car_lines = [f"{name} | {price} | {km}" for name, price, km in cars]
    return "\n".join(car_lines)
    #return str(car_lines)
```

```python
res = divar()
print(res)
```

```python
agent = Agent(
    name = "car buy assistant",
    markdown = True,
    model = llm,
    tools = [divar]
)
```

```python
response = agent.run('give me a list of cars with prices, names and km in nice and understanding format.')
print(response)
```
