```python
from langchain_groq import ChatGroq
from langchain_community.document_loaders import WebBaseLoader
from langchain.prompts import PromptTemplate
from langchain_core.output_parsers import JsonOutputParser
import pandas as pd
import chromadb
import uuid
```

```python
loader = WebBaseLoader("https://microsoft.ai/job/member-of-technical-staff-ai-personality-personalization/")
page_data = loader.load().pop().page_content
```

```python
prompt_extract = PromptTemplate.from_template(
    """
    متن کپی شده از وبسایت:
    {page_content}
    دستورالعمل:
    متن کپی شده، متنی از صفحه شغلی در یک وبسایت است.
    وظیفه شما استخراج آگهی های شغلی و بازگرداندن آن ها در قالب json، حاوی کلیدهای زیر است:
    'role', 'skill', 'experience', 'description'.
    فقط json معتبر را برگردانید.
    json معتبر، json ای است که بدون مقدمه باشد.
    توجه داشته باشید که مقادیر کلیدها به زبان فارسی ترجمه شوند و بازگردانده شوند.
    """
)

chain_extract = prompt_extract | llm
response = chain_extract.invoke(input = {'page_content': page_data})
print(response.content)
```

```python
json_parser = JsonOutputParser()
json_response = json_parser.parse(response.content)
json_response
```

```python
skills = ["پردازش زبان طبیعی", "پردازش زبان طبیعی", "یادگیری ماشین", "یادگیری ماشین", "بینایی ماشین"]
links = ["https://github.com/alikhademikhoubani/Neutral-Language-Processing/blob/main/Airline%20Reviews%20Sentiment%20Classification.ipynb",
        "https://github.com/alikhademikhoubani/Neutral-Language-Processing/blob/main/Article%20Summarization%20and%20Classification.ipynb",
        "https://github.com/alikhademikhoubani/Classification/blob/main/AdaBoostClassifier%201.ipynb",
        "https://github.com/alikhademikhoubani/Classification/blob/main/Airline%20Quality%20Classification.ipynb",
        "https://github.com/alikhademikhoubani/Computer-Vision/blob/main/Alzheimer's%20diagnosis.ipynb"]
df = pd.DataFrame()
df["skills"] = skills
df["links"] = links
df.head()
```

```python
client = chromadb.PersistentClient("vectorstore")
collection = client.get_or_create_collection(name = "temp")

if not collection.count():
    for _, row in df.iterrows():
        collection.add(documents = row["skills"],
                      metadatas = {"link": row["links"]},
                      ids = [str(uuid.uuid4())])
```

```python
links = collection.query(query_texts = ["پردازش زبان طبیعی", "یادگیری ماشین"], n_results = 2).get("metadatas")
links
```

```python
prompt_email = PromptTemplate.from_template(
    """
    شرح شغل:
    {job_description}
    دستورالعمل:
    شما یک مدیر کسب و کار در یک شرکت هستید. شرکت شما یک شرکت مشاوره هوش مصنوعی و نرم افزار است.
    وظیفه شما نوشتن یک ایمیل به مشتری در مورد شغل ذکر شده در بالا است که توانایی شرکت را در برآورده کردن نیازهای آن ها شرح می دهد.
    همچنین، مرتبط ترین موارد را از لینک های زیر برای نمایش نمونه کارهای شرکت اضافه کنید:
    {link_list}
    به یاد داشته باشید که شما مدیر کسب و کار در شرکت هستید.
    """
)

chain_email = prompt_email | llm
response = chain_email.invoke({"job_description": json_response["description"], "link_list": links})
print(response.content)
```
