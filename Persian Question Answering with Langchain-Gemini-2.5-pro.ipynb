```python
loader = CSVLoader(file_path = "/kaggle/working/outputs/my_dataframe.csv", source_column = "question")
data = loader.load()
```

```python
data
```

```python
def text_cleaning(text):
    if not isinstance(text, str):
        return text
    text = re.sub("\u200c", " ", text)  
    text = re.sub(r"\x0c", " ", text)   
    return text.strip()
```

```python
for doc in data:
    doc.page_content = text_cleaning(doc.page_content)
    doc.metadata = {
        key: text_cleaning(value) for key, value in doc.metadata.items()
    }
```

```python
data[:10]
```

```python
embeddings = HuggingFaceEmbeddings(model_name = "intfloat/multilingual-e5-base")
# for english texts:
# HuggingFaceInstructEmbeddings() -> should be empty as a default
```

```python
vectordb = FAISS.from_documents(documents = data, embedding = embeddings)
```

```python
retriever = vectordb.as_retriever()
```

```python
retriever.get_relevant_documents("میزوپرستول چه تاثیری روی معده می گذارد؟")
```

```python
retriever.get_relevant_documents("چرا اسهال چرب ایجاد می شود؟")
```

```python
chain = RetrievalQA.from_chain_type(
    llm = llm,
    chain_type = "stuff",
    retriever = retriever,
    input_key = "query",
    return_source_documents = True
)
```
