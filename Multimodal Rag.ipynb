```python
from pathlib import Path
import yt_dlp

def download_video(video_url: str, output_dir: str) -> str:
    output_dir = Path(output_dir)
    output_dir.mkdir(parents=True, exist_ok=True)

    ydl_opts = {
        "outtmpl": str(output_dir / "input_vid.%(ext)s"),
        "format": "bestvideo+bestaudio/best",
        "merge_output_format": "mp4",
        "noplaylist": True,
        "cookies_from_browser": ("chrome",),
        "progress_hooks": [lambda d: print(f"Status: {d['status']}")],
    }

    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        info_dict = ydl.extract_info(video_url, download=True)
        filename = ydl.prepare_filename(info_dict)
        if ydl_opts.get("merge_output_format"):
            filename = Path(filename).with_suffix(f".{ydl_opts['merge_output_format']}")

    return str(filename)

video_url = "https://youtu.be/3dhcmeOTZ_Q"
output_path = r"/content/content"

video_file = download_video(video_url, output_path)
print("Downloaded:", video_file)
```

```python
def video_to_images(video_path, output_folder):
    clip = VideoFileClip(video_path)
    clip.write_images_sequence(
        os.path.join(output_folder, "frame%04d.png"), fps = 0.2
    )
```

```python
def video_to_audio(video_path, output_audio_path):
    clip = VideoFileClip(video_path)
    audio = clip.audio
    audio.write_audiofile(output_audio_path)
```

```python
def audio_to_text(audio_path):
    recognizer = sr.Recognizer()
    audio = sr.AudioFile(audio_path)

    with audio as source:
        audio_data = recognizer.record(source)

        try:
            text = recognizer.recognize_whisper(audio_data)
        except sr.UnknownValueError:
            print("Speech recognition could not understand the audio.")
    return text
```

```python
import open_clip
import torch
from PIL import Image

model, _, preprocess = open_clip.create_model_and_transforms(
    "ViT-B-32", pretrained="openai"
)
tokenizer = open_clip.get_tokenizer("ViT-B-32")

def image_embedding(image_path):
    image = preprocess(Image.open(image_path)).unsqueeze(0)
    with torch.no_grad():
        emb = model.encode_image(image)
    return emb[0].cpu().numpy()
```

```python
retriever_engine = index.as_retriever(
    similarity_top_k=1,
    image_similarity_top_k=3
)
```

```python
def retrieve(retriever_engine, query_str):
    retrieval_results = retriever_engine.retrieve(query_str)

    retrieved_image = []
    retrieved_text = []
    for res_node in retrieval_results:
        if isinstance(res_node.node, ImageNode):
            retrieved_image.append(res_node.node.metadata["file_path"])
        else:
            display_source_node(res_node, source_length = 200)
            retrieved_text.append(res_node.text)

    return retrieved_image, retrieved_text
```

```python
def plot_images(images_path):
    images_shown = 0
    plt.figure(figsize = (16, 9))
    for img_path in images_path:
        if os.path.isfile(img_path):
            image = Image.open(img_path)

            plt.subplot(2, 3, images_shown + 1)
            plt.imshow(image)
            plt.xticks([])
            plt.yticks([])

            images_shown += 1
            if images_shown >= 5:
                break
```

```python
query = "what is linear regression explain to me?"
img, txt = retrieve(retriever_engine, query)
plot_images(img)
```

```python
from transformers import LlavaProcessor, LlavaForConditionalGeneration
import torch
from PIL import Image

model_id = "llava-hf/llava-1.5-7b-hf"

processor = LlavaProcessor.from_pretrained(model_id)
model = LlavaForConditionalGeneration.from_pretrained(
    model_id,
    torch_dtype=torch.float16,
    device_map="auto"
)
```

```python
def multimodal_answer(query, context, image_paths):
    images = [Image.open(p).convert("RGB") for p in image_paths[:3]]

    prompt = f"""
    Use the following context extracted from a video to answer the question.

    Context:
    {context}

    Question:
    {query}
    """

    inputs = processor(
        text=prompt,
        images=images,
        return_tensors="pt"
    ).to(model.device)

    output = model.generate(
        **inputs,
        max_new_tokens=300
    )

    return processor.decode(output[0], skip_special_tokens=True)
```

```python
def multimodal_answer(query, context_str, image_paths):
    images_to_use = [
        Image.open(image_paths[0]).convert("RGB").resize((224, 224))
        for p in image_paths[:1] if os.path.isfile(p)
    ]

    prompt = f"""
Use the following context extracted from a video to answer the question.
<image>

Context:
{context_str}

Question:
{query}
"""

    processor_kwargs = {"text": prompt, "return_tensors": "pt"}
    if len(images_to_use) > 0:
        processor_kwargs["images"] = images_to_use

    inputs = processor(**processor_kwargs).to(model.device)
    output = model.generate(**inputs, max_new_tokens=300)
    answer = processor.decode(output[0], skip_special_tokens=True)
    return answer
```
